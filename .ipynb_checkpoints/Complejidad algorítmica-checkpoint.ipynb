# Introducción








La complejidad algorítmica nos permite comparar 
la eficiencia entre algoritmos, y así predecir el tiempo en qué nos tardará ejecutar dicho algoritmo. 

La complejidad no solo es temporal, sino que también existe la complejidad espacial(espacio en memoria que necesitamos para poder resolver el problema). 

La complejidad algorítmica temporal en general se denota como T(n). Algunas aproximaciones para determinar esta complejidad algoritmica son:

- Cronometrar el tiempo de ejecución de un algoritmo.(nada confiable
- Contar los pasos de los procesos dentro del algoritmo.(medio confiable
- Contar los pasos conforme nos aproximamos al infinito.(confiable)

Formalmente se deberá tener en cuenta la última (modela el comportamiento),pues los tiempos de ejecución no son medida equitativa y varía respecto a las herramientas de computo, y el número de pasos no determina sino la estética del algoritmo en muchos casos. 



```python
import time 

def factorial(n):
        respuesta = 1
        
        while n>1:
            respuesta *=n
            n-=1
            
        return respuesta
    
def factorial_r(n):
         if n==1:
         
            return 1
        
         return n*factorial(n-1)
    

    
n = 200000

comienzo = time.time()
factorial(n)
fin= time.time()
print(fin-comienzo)

comienzo = time.time()
factorial_r(n)
fin= time.time()
print(fin-comienzo)
```

# Conteo Abstracto de operación

Es un método para determinar el número de veces que se ejecuta
una operación dentro de un aloritmo, para así determinar, posteriormente una medida para clasificar la eficiencia del algoritmo.

# Notación asintótica

### Big-O notation, Theta notation and Omega notation.

Aquí debemos tener en cuenta el crecimiento asintótico que tiene algunos hechos fundamentales como:

- No importa las pequeñas variaciones
- Se enfoca conforme la variable de análisis tiende al infinito.
- Se acota y se puede pensar en el "mejor de los casos", " caso promedio" o el "peor de los casos"
- Lo que más importa es el término de mayor tamaño(Big O notation - "peor de los casos")

Más información [aquí](https://www.programiz.com/dsa/asymptotic-notations) o en el libro "introduction to algorithms" MIT Press

Ley de la suma: crecimiento lineal

Ley de la multiplicación: crecimiento cuadrático

Recursividad múltiple: crecimiento exponencial

# Clases de complejidad algorítmica

- O(1) Constante
- O(n) Lineal
- O(log n) Logaritmica
- O(n log n) logarítmica Lineal
- O(n^2) Polinomial
- O(2^n) Exponencial
